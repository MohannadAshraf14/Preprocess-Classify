```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/Telco customer.csv')
```

```python
df.drop(['customerID','gender','PaymentMethod'],axis=1,inplace=True)
```

```python
df['TotalCharges']=pd.to_numeric(df['TotalCharges'],errors='coerce')
df['TotalCharges']=df['TotalCharges'].fillna(df['MonthlyCharges']*df['tenure'])
```

```python
df = df.replace({'No internet service':0,'No':0,'Yes':1})
df['MultipleLines']=df['MultipleLines'].replace({'Yes':1,'No':0,'No phone service':0})
df['InternetService']=df['InternetService'].replace({'Fiber optic':1,'DSL':2,'No':0})
df['Contract']=df['Contract'].replace({'Month-to-month':1,'One year':2,'Two year':3})
```

```python
x= df.drop('Churn',axis=1)
y= df['Churn']
```

```python
x.info()
```

```python
x.head()
```

```python
#smote
from imblearn.over_sampling import SMOTE
smote = SMOTE()
x,y = smote.fit_resample(x,y)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x = sc.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
reg = lr.fit(x_train, y_train)
```

```python
y_pred = reg.predict(x_test)
```

```python
reg.score(x_test,y_test)
```

```python
from sklearn.metrics import classification_report
y_pred = reg.predict(x_test)
print(classification_report(y_test, y_pred))
```

```python
model = LogisticRegression(multi_class='multinomial')
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
model.score(x_test,y_test)
```

```python
from sklearn.metrics import classification_report
y_pred = model.predict(x_test)
print(classification_report(y_test, y_pred))
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=5)
model=knn.fit(x_train, y_train)
```

```python
y_pred = knn.predict(x_test)
```

```python
model.score(x_test,y_test)
```

```python
from sklearn.metrics import classification_report
y_pred = model.predict(x_test)
print(classification_report(y_test, y_pred))
```

```python
from sklearn.svm import SVC
svm = SVC()
model=svm.fit(x_train, y_train)
```

```python
y_pred = svm.predict(x_test)
```
